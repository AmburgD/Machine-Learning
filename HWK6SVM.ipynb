```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('car_data.csv')
X = dataset.iloc[:, 2:-1].values
Y = dataset.iloc[:, -1].values
```

```python
print(dataset)
```

```python
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = .20, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = "rbf", random_state = 0)
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1),Y_test.reshape(len(Y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
print(X_test)
print(Y_test)
print(Y_pred)
```

```python
print(classifier.predict(sc.transform([[30, 50000]])))
```

```python
# create a mesh to plot in
h = 10  # step size in the mesh
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# predict the class labels for each point in the mesh
Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# plot the results
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.Set1, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Set1)
plt.xlabel('X feature 1')
plt.ylabel('X feature 2')
plt.title('SVM classification')
plt.show()
```

<h3> In this example I imported a dataset that predicted wether a person would choose to buy a car based on identifiers such as their income, age, and gender. I first removed all of the dummy variables. I then split the data, followed by scaling it. Next I implemented the SVM algorithm using the Gaussian kernel. When being tested using the other kernels as well as testing to see the accuracy if a linear method was used, the accuracy was the higest for the Gaussian kernel. This means that this model is a non-linear SVM. I created a prediction variable so that way I could build the confusion matrix. I then finally created an accuracy score and created a new data point to test the algorithm with.The data represents how likely a person is to purchase a car using classification.
