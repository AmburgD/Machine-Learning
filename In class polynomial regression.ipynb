```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1].values
```

```python
print(X)
print(Y)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, Y)
```

```python
## we used the equation for the polynomial regression... degree in this case means that we are using
## the fourth level polynomial in the equation (b4x^4 'th term)
## in order to determine the number of degrees, you guess and check to see what makes the line of 
## best fit. there is not much difference between 4th degree and 5th degree which makes that the cap
## essentially. 
```

```python
plt.scatter(X,Y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X,Y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Linear regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
## this is known as overfitting because we didnt split the data into training and testing. Therefore
## the code knows everything about the data and cannot recieve anything new from the data
```

```python
lin_reg.predict([[6.5]])
```

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```

```python
print(lin_reg_2.coef_)
print(lin_reg_2.intercept_)
```
